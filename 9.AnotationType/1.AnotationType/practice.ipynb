```python

from typing import Generator, TypeVar

NumInt = TypeVar('NumInt', bound=float)



def add_nums() -> Generator[int, NumInt, tuple[int, list[NumInt]]]:
    all_inputs: list[NumInt] = []
    total = 0
    while True:
        num: NumInt = yield len(all_inputs)
        if num is None:
            return total, all_inputs
        all_inputs.append(num)
        total += num

def proccess_nums(numbers: tuple[int, ...]):
    g = add_nums()
    print("init", next(g))
    g.send(2)
    g.send(2.5)
    print("got", g.send(1))

    for num in numbers:
        g.send(num)

    result = yield from g
    total, items = result
    yield total

print(add_nums())
f = add_nums()
for i in range(10):
    print(next(f))
```

```python
from typing import Generic, TypeVar

# Объявляем переменную типа (TypeVar), которая может быть абсолютно любым типом.
T = TypeVar("T")


class Queue(Generic[T]):
    def __init__(self) -> None:
        # Инициализируем пустой список для хранения элементов типа T
        self._items: list[T] = []

    def enqueue(self, item: T) -> None:
        """Добавляет элемент в конец очереди."""
        self._items.append(item)

    def dequeue(self) -> T:
        """Извлекает и возвращает элемент из начала очереди."""
        if self.is_empty():
            raise IndexError("Попытка извлечь элемент из пустой очереди")
        return self._items.pop(0)

    def is_empty(self) -> bool:
        """Проверяет, пуста ли очередь."""
        return len(self._items) == 0

    def __str__(self) -> str:
        return f"Queue({self._items})"



# 1. Создаем строковую очередь. Теперь везде внутри этого экземпляра T становится str.
str_queue = Queue[str]()

str_queue.enqueue("Первый")
str_queue.enqueue("Второй")

# Корректно: извлекаемый элемент автоматически имеет тип str
first_item: str = str_queue.dequeue()
print(f"Извлечено: {first_item}")  # Выведет: Первый

# 2. ТЕСТ НА ОШИБКУ ТИПИЗАЦИИ
# Если раскомментировать строку ниже, Python выполнит её (в рантайме это просто список)
# Но 2.mypy или PyCharm сразу подцветят её красным:
# "Expected type 'str', got 'int' instead"
# str_queue.enqueue(123)

# 3. Создаем очередь для чисел
int_queue = Queue[int]()
int_queue.enqueue(10)
int_queue.enqueue(20)

# IDE знает, что next_num — это int, и подсветит синтаксис, если попробовать вызвать метод строк
next_num = int_queue.dequeue()
# next_num.upper()  # Ошибка 2.mypy: у int нет метода upper()


```
