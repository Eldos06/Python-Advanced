```python
class A:
    pass

class B:
    pass

```

```python
a = A()
b = B()
```

```python


import sys

print(sys.getrefcount(a))
print(sys.getrefcount(b))
```

```python
a.b = b
print(sys.getrefcount(a))
print(sys.getrefcount(b))
```

```python
print(sys.getrefcount(a.b))
```

```python
b.a = a # strong links

print(sys.getrefcount(a))
print(sys.getrefcount(b))
```

```python
import weakref

b = B()
print(b)
print(sys.getrefcount(b))
```

```python
weak_ref_to_b = weakref.ref(b)
print(weak_ref_to_b)
print(sys.getrefcount(b))
```

```python
a.b = weak_ref_to_b
print(sys.getrefcount(b))
```

```python
print(a.b)
print(a.b())
```

```python
print(sys.getrefcount(b))
```

```python
del b
print(a.b())
```

```python
# dict with weak links
d = weakref.WeakKeyDictionary()
print(d)
```

```python
print(sys.getrefcount(a))

d[a] = "this is 'a'" # weak link
print(d.items())
print(dict(d))
```

```python
print(sys.getrefcount(a))
```

```python
b = B()
print(sys.getrefcount(b))
d[b] = "this is 'b'"
print(sys.getrefcount(b))
print(d[b])
print(dict(d))

```

```python
del b
print(dict(d))
```

```python
some_cache = weakref.WeakKeyDictionary()

def helper(obj):
    some_cache[obj] = "some cached obj data"
    b = B()
    some_cache[b] = "this is 'b-data'"
    print(sys.getrefcount(obj))
    data_obj =some_cache[obj]
    print("obj data", data_obj)
    data_b = some_cache[b]
    print("b data", data_b)
    print("some_cache", dict(some_cache))
    return data_obj

```

```python
a3 = A()
helper(a3)
print(dict(some_cache))
helper(a3)
print(dict(some_cache))
```

```python
import sys

name = 'Yeldos'
print(sys.getrefcount(name)) # Вернет N (больше 1 из-за интернирования строк и аргумента функции)

names = [name, "Dosic"]
# Счетчик `name` увеличился, так как список теперь тоже ссылается на него.
print(sys.getrefcount(name))

del names # Удаляем список -> счетчик name уменьшается обратно
```
